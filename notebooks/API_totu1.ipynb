```python
import requests
```

```python
import xmltodict
```

```python
resp = requests.get("http://odp.es.govt.nz/ariver.hts?Service=Hilltop&Request=GetData&site=Mataura%20River%20at%20Gore&measurement=Flow&From=23/3/2026&To=24/3/2026")
```

```python
resp.status_code
```

```python
the_data = xmltodict.parse(resp.content)
```

```python
the_data
```

```python

now_query = "http://odp.es.govt.nz/ariver.hts?Service=Hilltop&Request=GetData&site=Mataura%20River%20at%20Gore&measurement=Flow&From={from_date}&To={to_date}"

nowresp = requests.get(now_query.format(from_date="26/6/2026", to_date="27/6/2026"))
nowresp.raise_for_status()

now_data = xmltodict.parse(nowresp.content)
now_data
```
